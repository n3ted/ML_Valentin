# Simple linear regression 

Linear reg is a **supervised** learning algorithm, the predicted output is **continous**. When we have one feature or predictor var it's called **simple linear reg**, and with multiple features it's called **multiple lin reg**.
Equation for the **simple lin reg** is:

$y \approx \beta_0 + \beta_1X$,

where $X$ is the $predictor var$, and $y$ is the $respons var$.
$\beta_0$ is the $intercept$ and $\beta_1$ is the $slope$. **beta_0 and beta_1** are unknown parameters that needs to be **estimated** with the **datapoint** $(x_1, y_1), (x_2, y_2), \ldots, (x_n,y_n)$.

Example: 

- $y$ sales 
- $X$ - Tv spendings as it looks to fit the a line more then the others, through visually inspecting the **scatterplots**. 

We use the data points to get the sample estimate $\hat{\beta}_0$, $\hat{j\beta}_1$ by using the $OLS$ least square estimate, which is the most common method. Then we get the prediction $\hat{y} = \hat{\beta}_0+\hat{\beta}_1x$, which is a line that is as close as possible to the data points.

```python
X, y = df['TV'], df['Sales'] # create a tuple and TV goes into X and Sales col into y

# fits a polynomial of degree 'deg' using least squres polynomial fit
beta_1, beta_0 = np.polyfit(X, y, deg=1) # returns coeff with the highest power first

# predicted y, note that the beta_0 and beta_1 are acctually **hats** as they are sample estimates

y_hat = lambda x: beta_0 + beta_1 * x

print(f'Intercept beta_0 hat:{beta_0:.4f}')
print(f'Slope beta_1 hat:{beta_1:.4f}') # Spending 1000 $ gives 47.5 extra sold units
```

```python
spend = np.linspace(0, 350)
fig, ax = plt.figure(figsize=(5, 3), dpi=100), plt.axes()

sns.scatterplot(data=df, x='TV', y='Sales') # Create the scatter plot with tv on x-axis and sales on y-axis
sns.lineplot(x=spend, y=y_hat(spend), color='red') # Create the linear reg in the grapf

ax.set(
    title='TV advertisement linear regression',
    xlabel='Thousands dollars', 
    ylabel='Sales thousands of units'
)
```

## Multiple linear regression

We used numpy polyfit to estimate $\hat\beta_0, \hat\beta_1$ from the TV-feature alone. polyfit requires 1D vector, but we want to use all the features (TV, radio, newspaper) for prediction, i.e. we want to use more explanatory variables. One way is to manually solve the **normal equation** or the **closed form equation**: $\hat{\bf{\beta}} = (\bf{X}^T\bf{X})^{-1}\bf{X}^T\bf{y}$ as we did [linear algebra course](https://github.com/kokchun/Linjar-algebra-21/blob/main/Lectures/Lec6-Matrix/Linear_regression.ipynb), but now we add more features which is simply more columns.

where $\bf{X} = 	\begin{bmatrix} 
	1 & x_1^{(1)} & x_2^{(1)}& \ldots &x_n^{(1)} \\
	1 & x_1^{(2)} & x_2^{(2)}& \ldots &x_n^{(2)}\\
	\vdots & \vdots & \vdots & \ddots & \vdots\\
	1 & x_1^{(m)}& x_2^{(m)}& \ldots &x_n^{(m)}
\end{bmatrix}, \bf{y} = \begin{bmatrix} 
y_1 \\ y_2 \\ \vdots \\ y_m
\end{bmatrix}, \hat{\bf{\beta}} = \begin{bmatrix} 
\hat\beta_0 \\\hat\beta_1\\ \vdots \\ \hat\beta_n
\end{bmatrix}$

In our example we have $n = 3$ features, $m = 200$ samples, which gives us the regression coefficients $\hat\beta_0, \hat\beta_1, \hat\beta_2, \hat\beta_3$.

The regression line is $y = \hat\beta_0 + \hat\beta_1x_1 + \hat\beta_2x_2 + \hat\beta_3x_3$. With this equation we can predict the sale for a new sample $i$ by $\hat{y}^{(i)} = \hat\beta_0 + \hat\beta_1x_1^{(i)} + \hat\beta_2x_2^{(i)} + \hat\beta_3x_3^{(i)}$ or using dot product $\hat{y}_i = \hat{\beta}\cdot\bf{x}^{(i)}$, where $\bf{x}^{(i)}$ = $[1, x_1^{(i)}, x_2^{(i)},x_3^{(i)} ]^T$

```python
X, y = df.drop('Sales', axis='columns'), df['Sales'] 
''' Create a tuple X, y then drop from df the Sales col
and assign Sales col to y'''
X.insert(0,'Intercept', 1)

# normal eq, note that we use pinv, which stands for pseudoinverse instead for inverse 
# ..which uses something called SVD- Singular  Value Decomposition. This gives computational 
# ..complexity of 0(n^2) instead of 0(n^3)
reg_fit = lambda X, y: np.linalg.pinv(X.T @ X) @ X.T @ y # lambda arguments: expression
beta_hat = reg_fit(X, y) # Create approx for beta_hats
'''(np.linalg.pinv(X.T @ X) @ X.T @ y) calculates the coefficients for linear regression using the 
pseudoinverse. So, when you call reg_fit(X, y), it's like calling a regular function with X and y as
 inputs, and the function performs the specified linear regression calculation, returning the coefficients.'''
predict = lambda x, beta: np.dot(x, beta) 

# note that we add 1 to account for the the intercept
test_sample = [1, 230.1, 37.8, 69.2]

y_hat = predict(test_sample, beta_hat) #calculate the y hat by calling the predict fuction and send in
# ..test sample and beta_hat
print(beta_hat)
print(f'TV: $230.1k, Radio: $38.8k, Newspaper: $69.2k predicts {y_hat:.1f} thousands units')
```

# Train-test split

So how well did the model perform? There are several **evaluation metrics** that we can use to answer this question ***but not on the training data***. Using training data for evaluation we have ***data leakage***, because at the prediction time we shouldn't have this data available. **Data leakage** will lead to **overestimaton** of the performance as the model has already trained on the data it is using for eval. 

- We **split** the data in **a training set** and a **test set**, where the test-set will **only** be used during the **eval of the model**. Practicly we **randomly sample** this dataset without replacement with certain size for training set and rest for testing. 

```python
# split 70 % training and 30 test 
training_fraction = int(len(df) * 0.7)

# use sample method to get the fraction sample from df for trainingn, without replacement
train = df.sample(n=training_fraction, random_state=42, replace=False) 
test = df.drop(train.index) # the rest of data is the testset as we sampled without replacement

X_train, y_train = (
    train.drop('Sales', axis='columns'),
    train['Sales'],
)

X_test, y_test = test.drop(['Sales'], axis='columns'), test['Sales']
'''Create tuple then drop the Sales col and this goes int X_test
then the Sales col goes int y_test '''

# adding intercept column with ones 
X_train.insert(0,'Intercept', 1)
X_test.insert(0, 'Intercept', 1)

# checking shapes of the training|test split datasets
for dataset in ['X_train', 'X_test', 'y_train', 'y_test']:
    print(f'Shape of {dataset}, {eval(dataset).shape}')
```

## Evaluate performance

How well did we predict $\bf{y}$ with $\hat{\bf{y}}$?

To answer this question we use several **evaluation metrics** or **loss functions**: 

- Mean Absolute Error (MAE) - mean of error between $\bf{y}$ and ${\hat{\bf{y}}}$. The unit is same as measured quantity.

$$MAE = \frac{1}{m}\sum_{i=1}^m |y_i - \hat{y}_i|$$

- Mean Squared Error (MSE) - mean of squared errors between $\bf{y}$ and ${\hat{\bf{y}}}$. It punishes large errors, and the units are in square units of the measured quantity

$$MSE = \frac{1}{m}\sum_{i=1}^m (y_i - \hat{y}_i)^2$$

- Root Mean Squared Error (RMSE) - square root of MSE between $\bf{y}$ and ${\hat{\bf{y}}}$. It punishes large errors, and the units are same as measured quantity, hence easier to interpret.

$$RMSE = \sqrt{\frac{1}{m}\sum_{i=1}^m (y_i - \hat{y}_i)^2}$$


```python
m = len(y_test) # set m to len of y_test 
y_hat = np.reshape(y_hat, (m)) # removest the extra dimensions

MAE = 1 / m * np.sum(np.abs(y_test - y_hat))

MSE = 1 / m * np.sum((y_test - y_hat)**2)

RMSE = np.sqrt(MSE)

for metric in [
    ('Mean abs error', 'MAE', 'k units'),
    ('Mean square error', 'MSE', 'M units'),
    ('Root mean square error', 'RMSE', 'k units'),
]:
    print(f'{metric[0]}: {eval(metric[1]):.2f}{metric[2]}')
```
